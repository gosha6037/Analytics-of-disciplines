```python
import pandas as pd
import numpy as np
import ezodf
```

```python
def read_ods(filename, sheet_no=0, header=0):
    tab = ezodf.opendoc(filename=filename).sheets[sheet_no]
    return pd.DataFrame({col[header].value:[x.value for x in col[header+1:]]
                         for col in tab.columns()})
```

```python
import matplotlib.pyplot as plt
def HistOfDiscipline(title, color='blue'):
    plt.hist(df[title], bins=10, range=(0, 10), color=color)
    plt.title(title)
    plt.xlabel('Оценка')
    plt.ylabel('Количество человек')
    #plt.legend('title')
    plt.show()
```

```python
for column in df_2.columns[2:]:
    HistOfDiscipline(column, '#68228B')
```

```python
def NumberOfFailed(discipline):
    return df_2[df_2[discipline] < 4].shape[0]

FailedNumber = []
for title in df_2.columns[2:]:
    FailedNumber.append(NumberOfFailed(title))
labels = list(df_2.columns[2:])

new_label = []
new_FailedNumber = []
i=0
for item in FailedNumber:
    if item != 0:
        new_label.append(labels[i])
        new_FailedNumber.append(FailedNumber[i])
    i+=1
    
sizes = new_FailedNumber
explode = [0] * len(new_label)
explode[1] = 0.1
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=new_label, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
def NumberOfFailed(discipline):
    return df[df[discipline] < 4].shape[0]

FailedNumber = []
for title in df_2.columns[2:]:
    FailedNumber.append(NumberOfFailed(title))
labels = list(df_2.columns[2:])

new_label = []
new_FailedNumber = []
i=0
for item in FailedNumber:
    if item != 0:
        new_label.append(labels[i])
        new_FailedNumber.append(FailedNumber[i])
    i+=1
    
sizes = new_FailedNumber
explode = [0] * len(new_label)
for i in range(len(new_label)):
    if i % 2 == 1:
        explode[i] = 0.1

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=new_label, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
